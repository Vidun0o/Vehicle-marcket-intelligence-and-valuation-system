```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor

# 1. Load and Prep
df = pd.read_csv('car_price_dataset.csv')
df['Car_Age'] = 2025 - df['YOM']

# 2. Encode the 'Brand' column (Convert text to numbers)
le = LabelEncoder()
df['Brand_Encoded'] = le.fit_transform(df['Brand'])

# 3. Select Features including the Brand
X = df[['Brand_Encoded', 'Car_Age', 'Millage(KM)', 'Engine (cc)']]
y = df['Price']

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# 4. Predict (Fixing the Warning by using a DataFrame)
# Example: Predict price for an 'AUDI' (Brand 0), 5 years old, 50k KM, 1000cc
new_car = pd.DataFrame(
    [[0, 5, 50000, 1000]], 
    columns=['Brand_Encoded', 'Car_Age', 'Millage(KM)', 'Engine (cc)']
)

prediction = model.predict(new_car)
print(f"Predicted Price: {prediction[0]:.2f} Lakhs")
```
