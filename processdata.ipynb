```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('car_price_dataset.csv')


df['Car_Age'] = 2025 - df['YOM']


features = ['Price', 'Millage(KM)', 'Car_Age']
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])


kmeans = KMeans(n_clusters=4, random_state=42)
df['Market_Segment'] = kmeans.fit_predict(scaled_features)


df.to_csv('processed_car_data.csv', index=False)

print("Success! 'processed_car_data.csv' has been created in your folder.")
```
